|---|---|---|

| Liquidez Ampliada | Fondos Disp. / Depósitos | ≥ 14% |

```python
import duckdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.stattools import adfuller

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 5)
plt.rcParams['font.size'] = 11

DB  = r'C:\\Users\\Jonna\\OneDrive - St Paulinus Catholic Primary School\\claude proyectos\\IntelliCoop\SEPS_EEFF.duckdb'
RUC = '0691706710001'

con = duckdb.connect(DB, read_only=True)
print('Conexión exitosa ✓')
```

## 1. Extracción de cuentas clave

```python
# Cuentas del Plan Contable SFPS que necesitamos
CUENTAS_INTERES = {
    '1'   : 'activo_total',
    '2'   : 'pasivo_total',
    '3'   : 'patrimonio',
    '11'  : 'fondos_disponibles',
    '14'  : 'cartera_bruta',
    '1499': 'provisiones',
    '21'  : 'depositos',
    '4'   : 'gastos',
    '5'   : 'ingresos',
}

lista = "','" .join(CUENTAS_INTERES.keys())

raw = con.execute(f"""
    SELECT
        fecha_corte,
        cuenta,
        SUM(saldo_usd) AS saldo
    FROM eeff_mensuales
    WHERE ruc   = '{RUC}'
      AND cuenta IN ('{lista}')
    GROUP BY fecha_corte, cuenta
    ORDER BY fecha_corte, cuenta
""").df()

# Pivot: una fila por fecha, una columna por cuenta
df = raw.pivot(index='fecha_corte', columns='cuenta', values='saldo').reset_index()
df.columns.name = None
df = df.rename(columns=CUENTAS_INTERES)
df = df.sort_values('fecha_corte').reset_index(drop=True)

# Cartera improductiva: necesitamos cuentas 1425 (no devenga) + 1426 (vencida)
cart_imp = con.execute(f"""
    SELECT fecha_corte,
           SUM(CASE WHEN cuenta LIKE '1425%' THEN saldo_usd ELSE 0 END) AS no_devenga,
           SUM(CASE WHEN cuenta LIKE '1426%' THEN saldo_usd ELSE 0 END) AS vencida
    FROM eeff_mensuales
    WHERE ruc = '{RUC}'
    GROUP BY fecha_corte
    ORDER BY fecha_corte
""").df()

df = df.merge(cart_imp, on='fecha_corte', how='left')
df['cartera_improductiva'] = df['no_devenga'].fillna(0) + df['vencida'].fillna(0)

print(f'Períodos disponibles: {len(df)}')
print(f'Desde: {df.fecha_corte.min().date()}  →  Hasta: {df.fecha_corte.max().date()}')
df.head()
```

## 2. Cálculo de indicadores de solvencia

```python
# ── Indicadores ──────────────────────────────────────────────────
df['capitalizacion']   = df['patrimonio']  / df['activo_total']   * 100
df['apalancamiento']   = df['pasivo_total'] / df['patrimonio']
df['morosidad']        = np.where(
    df['cartera_bruta'] > 0,
    df['cartera_improductiva'] / df['cartera_bruta'] * 100, np.nan)
df['cobertura']        = np.where(
    df['cartera_improductiva'] > 0,
    df['provisiones'].abs() / df['cartera_improductiva'] * 100, np.nan)
df['resultado_neto']   = df['ingresos'].fillna(0) - df['gastos'].fillna(0)
df['roa']              = df['resultado_neto'] / df['activo_total'] * 100
df['liquidez_amplia']  = df['fondos_disponibles'] / df['depositos'] * 100

# ── Clasificación de riesgo por capitalización ───────────────────
def clasificar_solvencia(cap):
    if pd.isna(cap):      return 'Sin datos'
    elif cap >= 12:       return 'SOLVENTE'
    elif cap >= 9:        return 'ADECUADO'
    elif cap >= 6:        return 'ALERTA'
    else:                 return 'CRÍTICO'

df['estado_solvencia'] = df['capitalizacion'].apply(clasificar_solvencia)

COLORES_ESTADO = {'SOLVENTE':'#4CAF50','ADECUADO':'#8BC34A','ALERTA':'#FF9800','CRÍTICO':'#F44336','Sin datos':'#9E9E9E'}

# Tabla resumen
ultimo = df.iloc[-1]
print('=== ÚLTIMO CORTE DISPONIBLE:', ultimo['fecha_corte'].date(), '===')
print(f"  Capitalización:    {ultimo['capitalizacion']:.2f}%   (mín. 9%)")
print(f"  Morosidad:         {ultimo['morosidad']:.2f}%    (máx. 5%)")
print(f"  Cobertura cartera: {ultimo['cobertura']:.1f}%  (mín. 100%)")
print(f"  Apalancamiento:    {ultimo['apalancamiento']:.2f}x   (máx. 10x)")
print(f"  ROA:               {ultimo['roa']:.3f}%")
print(f"  Liquidez ampliada: {ultimo['liquidez_amplia']:.2f}%  (mín. 14%)")
print(f"  Estado solvencia:  {ultimo['estado_solvencia']}")
```

## 5. Modelo predictivo de solvencia — Holt-Winters + Regresión lineal
> Se proyecta el **Índice de Capitalización** a 12 meses con dos métodos

```python
# ── Preparar serie temporal ───────────────────────────────────────
serie_cap = (df[['fecha_corte','capitalizacion']]
             .dropna()
             .set_index('fecha_corte')
             .asfreq('ME', method='pad')  # frecuencia mensual
             ['capitalizacion'])

# ── Test de estacionariedad (ADF) ────────────────────────────────
adf_result = adfuller(serie_cap.dropna())
print(f'ADF Statistic: {adf_result[0]:.4f}')
print(f'p-value:       {adf_result[1]:.4f}')
print('Serie', 'ESTACIONARIA ✓' if adf_result[1] < 0.05 else 'NO estacionaria (tiene tendencia)')

# ── Horizonte de predicción ───────────────────────────────────────
H = 12  # meses hacia adelante
N = len(serie_cap)
fechas_pred = pd.date_range(start=serie_cap.index[-1] + pd.offsets.MonthEnd(1),
                             periods=H, freq='ME')

# ── Método 1: Regresión Lineal (tendencia) ───────────────────────
X = np.arange(N).reshape(-1,1)
y = serie_cap.values
reg = LinearRegression().fit(X, y)
X_fut = np.arange(N, N+H).reshape(-1,1)
pred_reg = reg.predict(X_fut)
r2 = reg.score(X, y)

# ── Método 2: Holt-Winters (suavizado exponencial con tendencia) ──
try:
    hw_model = ExponentialSmoothing(
        serie_cap, trend='add', seasonal=None, initialization_method='estimated'
    ).fit(optimized=True)
    pred_hw = hw_model.forecast(H).values
    hw_ok = True
except Exception as e:
    print('Holt-Winters no disponible:', e)
    pred_hw = pred_reg
    hw_ok = False

print(f'\nRegresión Lineal — R²: {r2:.4f}')
print(f'Pendiente mensual:    {reg.coef_[0]:.4f} pp/mes')
print(f'Tendencia anual:      {reg.coef_[0]*12:.3f} pp/año')
print(f'\nPredicción capitalización en 12 meses:')
print(f'  Regresión Lineal: {pred_reg[-1]:.2f}%')
if hw_ok:
    print(f'  Holt-Winters:     {pred_hw[-1]:.2f}%')
```

```python
# ── Gráfico de predicción ─────────────────────────────────────────
fig, ax = plt.subplots(figsize=(15, 6))

# Histórico
ax.plot(serie_cap.index, serie_cap.values, color='#1565C0', linewidth=2.5,
        label='Capitalización histórica', zorder=3)

# Regresión lineal completa (histórico + predicción)
X_all = np.arange(N+H).reshape(-1,1)
y_trend = reg.predict(X_all)
fechas_all = pd.date_range(start=serie_cap.index[0], periods=N+H, freq='ME')
ax.plot(fechas_all, y_trend, '--', color='#FF9800', linewidth=1.5,
        label=f'Tendencia lineal (R²={r2:.3f})', zorder=2)

# Predicción Holt-Winters
if hw_ok:
    ax.plot(fechas_pred, pred_hw, 'o-', color='#4CAF50', linewidth=2.5,
            markersize=5, label='Proyección Holt-Winters (12m)', zorder=4)
    # Banda de incertidumbre (±1.5 desv. estándar del error)
    residuos_std = (serie_cap.values - hw_model.fittedvalues.values).std()
    ax.fill_between(fechas_pred,
                    pred_hw - 1.96*residuos_std,
                    pred_hw + 1.96*residuos_std,
                    alpha=0.2, color='#4CAF50', label='IC 95%')

# Umbrales regulatorios
ax.axhline(9,  color='red',    linestyle='--', linewidth=1.2, alpha=0.7, label='Mínimo regulatorio 9%')
ax.axhline(12, color='orange', linestyle='--', linewidth=1.2, alpha=0.7, label='Óptimo 12%')

# Zonas de riesgo
ylim = ax.get_ylim()
ax.axvspan(serie_cap.index[-1], fechas_pred[-1], alpha=0.06, color='gray', label='Zona proyectada')

ax.set_title('Índice de Capitalización — Histórico y Proyección 12 meses\nCooperativa Fernando Daquilema',
             fontweight='bold', fontsize=12)
ax.set_ylabel('Capitalización (%)')
ax.legend(loc='upper left', fontsize=9)
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Semáforo de riesgo — estado actual vs umbrales SEPS

```python
ultimo = df.iloc[-1]

indicadores = [
    ('Capitalización',          ultimo['capitalizacion'],    9,   'mayor', '%'),
    ('Liquidez Ampliada',       ultimo['liquidez_amplia'],   14,  'mayor', '%'),
    ('Cobertura Cartera',       ultimo['cobertura'],         100, 'mayor', '%'),
    ('Morosidad NPL',           ultimo['morosidad'],         5,   'menor', '%'),
    ('Apalancamiento',          ultimo['apalancamiento'],    10,  'menor', 'x'),
    ('ROA',                     ultimo['roa'],               0,   'mayor', '%'),
]

def semaforo(valor, umbral, tipo):
    if pd.isna(valor): return '⚪ Sin datos', '#9E9E9E'
    ok = valor >= umbral if tipo == 'mayor' else valor <= umbral
    margen = abs(valor - umbral)
    if ok and margen >= abs(umbral)*0.3: return '🟢 SOLVENTE', '#4CAF50'
    elif ok:                             return '🟡 ADECUADO', '#FFC107'
    elif margen <= abs(umbral)*0.2:      return '🟠 ALERTA',   '#FF9800'
    else:                                return '🔴 CRÍTICO',  '#F44336'

print(f"{'Indicador':<28} {'Valor':>10} {'Umbral':>10} {'Estado':<20}")
print('─'*70)
for nombre, valor, umbral, tipo, unidad in indicadores:
    estado, _ = semaforo(valor, umbral, tipo)
    v_str = f'{valor:.2f}{unidad}' if not pd.isna(valor) else 'N/D'
    u_str = f'{umbral}{unidad}'
    print(f"{nombre:<28} {v_str:>10} {u_str:>10} {estado}")

# ── Panel visual tipo semáforo ────────────────────────────────────
fig, axes = plt.subplots(1, len(indicadores), figsize=(16, 4))
fig.suptitle(f'Semáforo de Riesgo — Fernando Daquilema ({ultimo["fecha_corte"].date()})',
             fontweight='bold', fontsize=13)

for ax, (nombre, valor, umbral, tipo, unidad) in zip(axes, indicadores):
    estado, color = semaforo(valor, umbral, tipo)
    ax.set_facecolor(color)
    ax.set_xlim(0,1); ax.set_ylim(0,1)
    ax.set_xticks([]); ax.set_yticks([])
    v_txt = f"{valor:.1f}{unidad}" if not pd.isna(valor) else 'N/D'
    ax.text(0.5, 0.62, v_txt, ha='center', va='center',
            fontsize=16, fontweight='bold', color='white')
    ax.text(0.5, 0.35, f'umbral {umbral}{unidad}', ha='center', va='center',
            fontsize=9, color='white', alpha=0.85)
    ax.set_title(nombre, fontsize=9, fontweight='bold', pad=4)

plt.tight_layout()
plt.show()
```
