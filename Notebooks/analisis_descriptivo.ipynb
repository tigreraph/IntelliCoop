```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Estilo de gráficos
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (13, 5)
plt.rcParams['font.size'] = 11

DB = r'C:\\Users\\Jonna\\OneDrive - St Paulinus Catholic Primary School\\claude proyectos\\IntelliCoop\SEPS_EEFF.duckdb'
con = duckdb.connect(DB, read_only=True)
print('Conexión exitosa ✓')
```

## 1. Vista general de la base de datos

```python
resumen = con.execute("""
    SELECT
        año,
        COUNT(DISTINCT fecha_corte)  AS meses,
        COUNT(DISTINCT ruc)          AS cooperativas,
        COUNT(*)                     AS registros
    FROM eeff_mensuales
    GROUP BY año
    ORDER BY año
""").df()

resumen['registros'] = resumen['registros'].apply(lambda x: f'{x:,.0f}')
print('=== REGISTROS POR AÑO ===')
display(resumen)
```

```python
# Segmentos disponibles
seg = con.execute("""
    SELECT segmento,
           COUNT(DISTINCT ruc) AS cooperativas
    FROM eeff_mensuales
    WHERE año = '2025'
    GROUP BY segmento ORDER BY segmento
""").df()

print('=== SEGMENTOS (año 2025) ===')
display(seg)
```

## 2. Evolución de cooperativas activas por año

```python
df_coop = con.execute("""
    SELECT año, segmento, COUNT(DISTINCT ruc) AS cooperativas
    FROM eeff_mensuales
    GROUP BY año, segmento
    ORDER BY año, segmento
""").df()

pivot = df_coop.pivot(index='año', columns='segmento', values='cooperativas').fillna(0)

pivot.plot(kind='bar', stacked=True, figsize=(13, 5), colormap='Set2')
plt.title('Cooperativas activas por segmento y año', fontweight='bold')
plt.xlabel('Año')
plt.ylabel('N° de cooperativas')
plt.xticks(rotation=45)
plt.legend(title='Segmento', bbox_to_anchor=(1.01, 1))
plt.tight_layout()
plt.show()
```

## 3. Activo Total del sistema — evolución anual
> Cuenta **1** = Activo

```python
activo = con.execute("""
    SELECT
        fecha_corte,
        año,
        segmento,
        SUM(saldo_usd) / 1e6 AS activo_millones
    FROM eeff_mensuales
    WHERE cuenta = '1'
    GROUP BY fecha_corte, año, segmento
    ORDER BY fecha_corte, segmento
""").df()

# Diciembre de cada año (o último mes disponible)
activo_dic = activo.sort_values('fecha_corte').groupby(['año','segmento']).last().reset_index()

pivot_a = activo_dic.pivot(index='año', columns='segmento', values='activo_millones').fillna(0)

ax = pivot_a.plot(kind='bar', stacked=True, figsize=(13, 5), colormap='Blues')
plt.title('Activo Total del sistema por segmento (último mes del año)', fontweight='bold')
plt.xlabel('Año')
plt.ylabel('Millones USD')
plt.xticks(rotation=45)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}M'))
plt.legend(title='Segmento', bbox_to_anchor=(1.01, 1))
plt.tight_layout()
plt.show()

print('\nActivo total por año y segmento (millones USD):')
display(pivot_a.round(1))
```

## 4. Liquidez — Fondos Disponibles (11) e Inversiones (13)

```python
liquidez = con.execute("""
    SELECT
        fecha_corte,
        año,
        CASE
            WHEN LEFT(cuenta,2) = '11' THEN 'Fondos Disponibles'
            WHEN LEFT(cuenta,2) = '13' THEN 'Inversiones'
        END AS tipo,
        SUM(saldo_usd) / 1e6 AS saldo_millones
    FROM eeff_mensuales
    WHERE LEFT(cuenta,2) IN ('11','13')
      AND LENGTH(cuenta) = 2
    GROUP BY fecha_corte, año, tipo
    ORDER BY fecha_corte, tipo
""").df()

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for ax, tipo in zip(axes, ['Fondos Disponibles', 'Inversiones']):
    sub = liquidez[liquidez['tipo'] == tipo]
    ax.plot(sub['fecha_corte'], sub['saldo_millones'], linewidth=2)
    ax.set_title(f'{tipo} — Sistema total', fontweight='bold')
    ax.set_ylabel('Millones USD')
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}M'))
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 5. Cartera de Crédito (14) — evolución

```python
cartera = con.execute("""
    SELECT
        fecha_corte,
        segmento,
        SUM(saldo_usd) / 1e6 AS cartera_millones
    FROM eeff_mensuales
    WHERE cuenta = '14'
    GROUP BY fecha_corte, segmento
    ORDER BY fecha_corte, segmento
""").df()

for seg in cartera['segmento'].unique():
    sub = cartera[cartera['segmento'] == seg]
    plt.plot(sub['fecha_corte'], sub['cartera_millones'], label=seg, linewidth=2)

plt.title('Cartera de Crédito por segmento', fontweight='bold')
plt.ylabel('Millones USD')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}M'))
plt.legend(title='Segmento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Depósitos (21) — evolución

```python
depositos = con.execute("""
    SELECT
        fecha_corte,
        segmento,
        SUM(saldo_usd) / 1e6 AS depositos_millones
    FROM eeff_mensuales
    WHERE cuenta = '21'
    GROUP BY fecha_corte, segmento
    ORDER BY fecha_corte, segmento
""").df()

for seg in depositos['segmento'].unique():
    sub = depositos[depositos['segmento'] == seg]
    plt.plot(sub['fecha_corte'], sub['depositos_millones'], label=seg, linewidth=2)

plt.title('Depósitos del público por segmento', fontweight='bold')
plt.ylabel('Millones USD')
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}M'))
plt.legend(title='Segmento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 7. Top 15 cooperativas — Activo Total (último dato disponible)

```python
ultima_fecha = con.execute("SELECT MAX(fecha_corte) FROM eeff_mensuales").fetchone()[0]
print(f'Último corte disponible: {ultima_fecha}')

top15 = con.execute(f"""
    SELECT
        razon_social,
        segmento,
        SUM(saldo_usd) / 1e6 AS activo_millones
    FROM eeff_mensuales
    WHERE cuenta = '1'
      AND fecha_corte = '{ultima_fecha}'
    GROUP BY razon_social, segmento
    ORDER BY activo_millones DESC
    LIMIT 15
""").df()

# Nombre corto
top15['nombre_corto'] = top15['razon_social'].str.replace(
    r'COOPERATIVA DE AHORRO Y CR[EÉ]DITO\s*', '', regex=True).str.strip().str.title()

colores = {'SEGMENTO 1': '#2196F3', 'SEGMENTO 1 MUTUALISTA': '#4CAF50',
           'SEGMENTO 2': '#FF9800', 'SEGMENTO 3': '#9C27B0'}
bar_colors = top15['segmento'].map(colores).fillna('#999')

fig, ax = plt.subplots(figsize=(13, 7))
bars = ax.barh(top15['nombre_corto'][::-1], top15['activo_millones'][::-1], color=bar_colors[::-1])
ax.set_xlabel('Millones USD')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}M'))
ax.set_title(f'Top 15 Cooperativas por Activo Total — {ultima_fecha}', fontweight='bold')

for bar, val in zip(bars, top15['activo_millones'][::-1]):
    ax.text(val + 5, bar.get_y() + bar.get_height()/2,
            f'${val:,.0f}M', va='center', fontsize=9)

# Leyenda segmento
from matplotlib.patches import Patch
leyenda = [Patch(color=c, label=s) for s, c in colores.items()]
ax.legend(handles=leyenda, title='Segmento', loc='lower right')
plt.tight_layout()
plt.show()

display(top15[['razon_social','segmento','activo_millones']].round(1))
```

## 8. Estadísticos descriptivos del Activo Total por segmento (2025)

```python
activos_2025 = con.execute("""
    SELECT
        ruc,
        razon_social,
        segmento,
        SUM(saldo_usd) / 1e6 AS activo_millones
    FROM eeff_mensuales
    WHERE cuenta = '1'
      AND año = '2025'
      AND MONTH(fecha_corte) = 12
    GROUP BY ruc, razon_social, segmento
    ORDER BY activo_millones DESC
""").df()

print('=== DESCRIPTIVOS POR SEGMENTO — Activo Total dic-2025 (millones USD) ===')
desc = activos_2025.groupby('segmento')['activo_millones'].describe().round(2)
display(desc)

# Boxplot
fig, ax = plt.subplots(figsize=(11, 5))
activos_2025.boxplot(column='activo_millones', by='segmento', ax=ax, showfliers=True)
ax.set_title('Distribución del Activo Total por segmento — Dic 2025', fontweight='bold')
ax.set_xlabel('Segmento')
ax.set_ylabel('Millones USD')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

## 9. Tabla resumen para análisis propio
Filtra lo que necesites cambiando los parámetros

```python
# ── Modifica estos parámetros según lo que quieras analizar ──
AÑO       = '2025'
SEGMENTO  = 'SEGMENTO 1'   # opciones: 'SEGMENTO 1', 'SEGMENTO 2', 'SEGMENTO 3'
CUENTAS   = ['1', '11', '13', '14', '2', '21', '3', '5', '4']  # grupos principales

tabla = con.execute(f"""
    SELECT
        fecha_corte,
        ruc,
        razon_social,
        cuenta,
        descripcion_cuenta,
        SUM(saldo_usd) / 1e6 AS saldo_millones
    FROM eeff_mensuales
    WHERE año      = '{AÑO}'
      AND segmento = '{SEGMENTO}'
      AND cuenta   IN ({','.join([repr(c) for c in CUENTAS])})
    GROUP BY fecha_corte, ruc, razon_social, cuenta, descripcion_cuenta
    ORDER BY fecha_corte, ruc, cuenta
""").df()

print(f'Filas: {len(tabla):,}  |  Año: {AÑO}  |  Segmento: {SEGMENTO}')
display(tabla.head(20))
```
