## 1. Extracción y Construcción del Panel de Indicadores

Se usa **diciembre 2025** como corte de análisis transversal (cierre anual completo).  
Para indicadores de flujo (ingresos, gastos) el acumulado a diciembre equivale al dato anual.

```python
# ── Cuentas de cartera improductiva (4 dígitos, sin subcuentas) ──────────────
# No Devenga Intereses: 1425–1448 | Vencida: 1449–1466
C_NODEV  = "'1425','1426','1427','1428','1432','1433','1434','1435','1436'," \
           "'1440','1441','1442','1443','1444','1448'"
C_VENC   = "'1449','1450','1451','1452','1456','1457','1458','1459'"

raw = con.execute(f"""
    SELECT
        ruc,
        razon_social,
        MAX(CASE WHEN cuenta='1'    THEN saldo_usd END)  AS activo_total,
        MAX(CASE WHEN cuenta='2'    THEN saldo_usd END)  AS pasivo_total,
        MAX(CASE WHEN cuenta='3'    THEN saldo_usd END)  AS patrimonio,
        MAX(CASE WHEN cuenta='11'   THEN saldo_usd END)  AS fondos_disp,
        MAX(CASE WHEN cuenta='13'   THEN saldo_usd END)  AS inversiones,
        MAX(CASE WHEN cuenta='14'   THEN saldo_usd END)  AS cartera_bruta,
        MAX(CASE WHEN cuenta='1499' THEN saldo_usd END)  AS provisiones,
        MAX(CASE WHEN cuenta='21'   THEN saldo_usd END)  AS depositos,
        MAX(CASE WHEN cuenta='26'   THEN saldo_usd END)  AS oblig_financ,
        MAX(CASE WHEN cuenta='31'   THEN saldo_usd END)  AS capital_social,
        MAX(CASE WHEN cuenta='33'   THEN saldo_usd END)  AS reservas,
        MAX(CASE WHEN cuenta='4'    THEN saldo_usd END)  AS gastos_total,
        MAX(CASE WHEN cuenta='41'   THEN saldo_usd END)  AS int_causados,
        MAX(CASE WHEN cuenta='44'   THEN saldo_usd END)  AS prov_gasto,
        MAX(CASE WHEN cuenta='45'   THEN saldo_usd END)  AS gastos_op,
        MAX(CASE WHEN cuenta='5'    THEN saldo_usd END)  AS ingresos_total,
        MAX(CASE WHEN cuenta='51'   THEN saldo_usd END)  AS int_ganados,
        SUM(CASE WHEN cuenta IN ({C_NODEV}) THEN saldo_usd ELSE 0 END) AS cart_nodev,
        SUM(CASE WHEN cuenta IN ({C_VENC})  THEN saldo_usd ELSE 0 END) AS cart_vencida
    FROM eeff_mensuales
    WHERE segmento    = 'SEGMENTO 1'
      AND fecha_corte = '2025-12-31'
      AND razon_social NOT LIKE '%LIQUIDACI%'
    GROUP BY ruc, razon_social
""").df()

print(f'Cooperativas Segmento 1 — dic-2025: {len(raw)}')
print(raw[['razon_social','activo_total']].sort_values('activo_total', ascending=False)
        .assign(activo_m=lambda d: (d.activo_total/1e6).round(1))
        [['razon_social','activo_m']].to_string(index=False))
```

## 2. Cálculo de los 15 Indicadores CAMEL/PERLAS

```python
df = raw.copy()

# ── Variables auxiliares ─────────────────────────────────────────────────────
df['resultado']         = df['ingresos_total'].fillna(0) - df['gastos_total'].fillna(0)
df['cart_improductiva'] = df['cart_nodev'].fillna(0) + df['cart_vencida'].fillna(0)
df['margen_fin_bruto']  = df['int_ganados'].fillna(0) - df['int_causados'].fillna(0)

# Nombre corto para gráficos
df['nombre'] = (df['razon_social']
                .str.replace(r'COOPERATIVA DE AHORRO Y CR[EÉ]DITO\s*', '', regex=True)
                .str.replace(r'LIMITADA|LTDA\.?', '', regex=True)
                .str.strip().str.title())

# ── CAMEL C — Capital ────────────────────────────────────────────────────────
df['C_solvencia']     = np.where(df['activo_total'] > 0,
                                  df['patrimonio'] / df['activo_total'] * 100, np.nan)

# ── CAMEL A — Calidad de Activos ─────────────────────────────────────────────
df['A_morosidad']     = np.where(df['cartera_bruta'] > 0,
                                  df['cart_improductiva'] / df['cartera_bruta'] * 100, np.nan)
df['A_cobertura']     = np.where(df['cart_improductiva'] > 0,
                                  df['provisiones'].abs() / df['cart_improductiva'] * 100, np.nan)
df['A_activos_prod']  = np.where(df['activo_total'] > 0,
                                  (df['cartera_bruta'].fillna(0) + df['inversiones'].fillna(0))
                                  / df['activo_total'] * 100, np.nan)

# ── CAMEL M — Gestión ────────────────────────────────────────────────────────
df['M_efic_operativa']= np.where(df['activo_total'] > 0,
                                  df['gastos_op'].fillna(0) / df['activo_total'] * 100, np.nan)
df['M_efic_ratio']    = np.where(df['margen_fin_bruto'] > 0,
                                  df['gastos_op'].fillna(0) / df['margen_fin_bruto'] * 100, np.nan)
df['M_intermediacion']= np.where(df['depositos'] > 0,
                                  df['cartera_bruta'] / df['depositos'] * 100, np.nan)

# ── CAMEL E — Rentabilidad ───────────────────────────────────────────────────
df['E_roa']           = np.where(df['activo_total'] > 0,
                                  df['resultado'] / df['activo_total'] * 100, np.nan)
df['E_roe']           = np.where(df['patrimonio'] > 0,
                                  df['resultado'] / df['patrimonio'] * 100, np.nan)
df['E_nim']           = np.where(df['activo_total'] > 0,
                                  df['margen_fin_bruto'] / df['activo_total'] * 100, np.nan)

# ── CAMEL L — Liquidez ───────────────────────────────────────────────────────
df['L_liq_inmediata'] = np.where(df['depositos'] > 0,
                                  df['fondos_disp'] / df['depositos'] * 100, np.nan)
df['L_liq_ampliada']  = np.where(df['depositos'] > 0,
                                  (df['fondos_disp'].fillna(0) + df['inversiones'].fillna(0))
                                  / df['depositos'] * 100, np.nan)

# ── S — Estructura ───────────────────────────────────────────────────────────
df['S_apalancamiento']= np.where(df['patrimonio'] > 0,
                                  df['pasivo_total'] / df['patrimonio'], np.nan)
df['S_cartera_activo']= np.where(df['activo_total'] > 0,
                                  df['cartera_bruta'] / df['activo_total'] * 100, np.nan)
df['S_aprovision']    = np.where(df['cartera_bruta'] > 0,
                                  df['provisiones'].abs() / df['cartera_bruta'] * 100, np.nan)

# ── Tamaño (control) ─────────────────────────────────────────────────────────
df['log_activo']      = np.where(df['activo_total'] > 0,
                                  np.log(df['activo_total']), np.nan)

RATIOS = ['C_solvencia','A_morosidad','A_cobertura','A_activos_prod',
          'M_efic_operativa','M_efic_ratio','M_intermediacion',
          'E_roa','E_roe','E_nim',
          'L_liq_inmediata','L_liq_ampliada',
          'S_apalancamiento','S_cartera_activo','S_aprovision']

print(f'Indicadores calculados: {len(RATIOS)}')
print(f'Observaciones válidas:  {df[RATIOS].notna().all(axis=1).sum()} / {len(df)}')
df[['nombre'] + RATIOS].set_index('nombre').round(2)
```

## 3. Estadísticos Descriptivos y Detección de Valores Atípicos

Se reportan mediana, media, desviación estándar, coeficiente de variación (CV), y los percentiles P10/P90.  
Los **outliers** se identifican con el criterio IQR (Tukey, 1977): $x < Q1 - 1.5 \cdot IQR$ o $x > Q3 + 1.5 \cdot IQR$.

```python
UMBRALES = {
    'C_solvencia':     ('mayor', 9,   '%'),
    'A_morosidad':     ('menor', 5,   '%'),
    'A_cobertura':     ('mayor', 100, '%'),
    'A_activos_prod':  ('mayor', 80,  '%'),
    'M_efic_operativa':('menor', 5,   '%'),
    'M_efic_ratio':    ('menor', 60,  '%'),
    'M_intermediacion':('mayor', 70,  '%'),
    'E_roa':           ('mayor', 1,   '%'),
    'E_roe':           ('mayor', 5,   '%'),
    'E_nim':           ('mayor', 3,   '%'),
    'L_liq_inmediata': ('mayor', 10,  '%'),
    'L_liq_ampliada':  ('mayor', 14,  '%'),
    'S_apalancamiento':('menor', 10,  'x'),
    'S_cartera_activo':('mayor', 60,  '%'),
    'S_aprovision':    ('mayor', 3,   '%'),
}

desc_rows = []
for r in RATIOS:
    s = df[r].dropna()
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    n_out = ((s < q1 - 1.5*iqr) | (s > q3 + 1.5*iqr)).sum()
    tipo, umbral, unit = UMBRALES[r]
    n_incumple = (s < umbral).sum() if tipo=='mayor' else (s > umbral).sum()
    desc_rows.append({
        'Indicador': r, 'N': len(s),
        'Media': s.mean(), 'Mediana': s.median(),
        'Std': s.std(), 'CV%': s.std()/s.mean()*100 if s.mean()!=0 else np.nan,
        'Min': s.min(), 'P10': s.quantile(0.10),
        'P90': s.quantile(0.90), 'Max': s.max(),
        'Umbral': f"{'≥' if tipo=='mayor' else '≤'}{umbral}{unit}",
        'N_incumple': n_incumple,
        'N_outliers_IQR': n_out
    })

desc_df = pd.DataFrame(desc_rows).set_index('Indicador')
pd.set_option('display.float_format', '{:.2f}'.format)
display(desc_df)
```

```python
fig, axes = plt.subplots(3, 5, figsize=(22, 14))
fig.suptitle('Distribución de Indicadores CAMEL — Segmento 1 | Dic 2025',
             fontsize=14, fontweight='bold', y=1.01)

for ax, ratio in zip(axes.flat, RATIOS):
    data = df[ratio].dropna()
    tipo, umbral, unit = UMBRALES[ratio]

    ax.boxplot(data, vert=True, patch_artist=True,
               boxprops=dict(facecolor='#90CAF9', color='#1565C0'),
               medianprops=dict(color='#C62828', linewidth=2),
               whiskerprops=dict(color='#1565C0'),
               capprops=dict(color='#1565C0'),
               flierprops=dict(marker='o', markerfacecolor='#F44336', markersize=5))

    ax.axhline(umbral, color='red', linestyle='--', linewidth=1.2,
               label=f'umbral {umbral}{unit}')

    # Jitter de puntos individuales
    jitter = np.random.uniform(-0.08, 0.08, len(data))
    ax.scatter(1 + jitter, data, alpha=0.5, s=18, color='#1565C0', zorder=3)

    ax.set_title(ratio, fontsize=9, fontweight='bold')
    ax.set_xticks([])
    ax.legend(fontsize=7, loc='upper right')

plt.tight_layout()
plt.savefig('fig1_boxplots_camel.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Análisis de Componentes Principales (PCA)

Se escalan los datos con **RobustScaler** (mediana y IQR) para atenuar el efecto de valores extremos antes de la PCA (Hubert & Vandervieren, 2008). Se reporta el biplot PC1–PC2 con loadings, varianza acumulada, y la contribución de cada variable.

```python
scaler  = RobustScaler()
X_sc    = scaler.fit_transform(df_clu[FEATURES_CLU])
nombres = df_clu['nombre'].values

pca = PCA(random_state=42)
pca.fit(X_sc)
Z   = pca.transform(X_sc)

var_exp = pca.explained_variance_ratio_
var_cum = np.cumsum(var_exp)

# ── Figura 1: Varianza explicada ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(range(1, len(var_exp)+1), var_exp*100, color='#1565C0', alpha=0.8)
axes[0].plot(range(1, len(var_exp)+1), var_cum*100, 'ro-', linewidth=2)
axes[0].axhline(70, color='gray', linestyle='--', alpha=0.7)
axes[0].set_xlabel('Componente Principal'); axes[0].set_ylabel('%')
axes[0].set_title('Varianza Explicada por Componente (Scree Plot)', fontweight='bold')
for i, (v, c) in enumerate(zip(var_exp, var_cum)):
    axes[0].text(i+1, v*100+0.5, f'{v*100:.1f}%', ha='center', fontsize=8)

# ── Figura 2: Loadings heatmap ───────────────────────────────────────────────
n_pc = np.argmax(var_cum >= 0.70) + 1
load_df = pd.DataFrame(pca.components_[:n_pc].T,
                        index=FEATURES_CLU,
                        columns=[f'PC{i+1}' for i in range(n_pc)])
sns.heatmap(load_df, cmap=sns.diverging_palette(220, 10, as_cmap=True),
            vmax=1, vmin=-1, center=0, annot=True, fmt='.2f',
            linewidths=.5, ax=axes[1])
axes[1].set_title(f'Loadings de las primeras {n_pc} CPs (≥70% varianza)', fontweight='bold')

plt.tight_layout()
plt.savefig('fig3_pca_varianza.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'\nComponentes necesarios para explicar ≥70% varianza: {n_pc}')
print(f'PC1: {var_exp[0]*100:.1f}%  |  PC2: {var_exp[1]*100:.1f}%  |  PC3: {var_exp[2]*100:.1f}%')
```

```python
fig, ax = plt.subplots(figsize=(14, 9))

ax.scatter(Z[:, 0], Z[:, 1], c='#1565C0', s=60, zorder=3, alpha=0.8)
for i, name in enumerate(nombres):
    ax.annotate(name, (Z[i, 0], Z[i, 1]),
                fontsize=7.5, ha='left', va='bottom',
                xytext=(3, 3), textcoords='offset points')

# Flechas de loadings (escaladas)
scale = 3.5
for j, feat in enumerate(FEATURES_CLU):
    ax.annotate('', xy=(pca.components_[0, j]*scale, pca.components_[1, j]*scale),
                xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='#C62828', lw=1.5))
    ax.text(pca.components_[0, j]*scale*1.08,
            pca.components_[1, j]*scale*1.08,
            feat, color='#C62828', fontsize=8, fontweight='bold')

ax.axhline(0, color='gray', linestyle='--', linewidth=0.7)
ax.axvline(0, color='gray', linestyle='--', linewidth=0.7)
ax.set_xlabel(f'PC1 ({var_exp[0]*100:.1f}% varianza explicada)', fontsize=11)
ax.set_ylabel(f'PC2 ({var_exp[1]*100:.1f}% varianza explicada)', fontsize=11)
ax.set_title('Biplot PCA — Cooperativas Segmento 1 (dic-2025)\n'
             'Flechas rojas = loadings de indicadores CAMEL; '
             'puntos azules = cooperativas', fontweight='bold')
plt.tight_layout()
plt.savefig('fig4_biplot_pca.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Análisis de Clustering

### 6.1 Determinación del número óptimo de clusters

Se aplican cuatro criterios de validación interna (Arbelaitz et al., 2013):

| Índice | Optimización | Referencia |
|---|---|---|
| **Silhouette** (Ŝ) | Maximizar (rango: −1 a 1) | Rousseeuw (1987) |
| **Calinski-Harabász** (CH) | Maximizar | Caliński & Harabaász (1974) |
| **Davies-Bouldin** (DB) | Minimizar | Davies & Bouldin (1979) |
| **Inercia (codo)** | Inflexión | Thorndike (1953) |

```python
from sklearn.cluster import KMeans

K_range = range(2, min(len(df_clu)-1, 9))
results = {'k': [], 'inertia': [], 'silhouette': [], 'ch': [], 'db': []}

np.random.seed(42)
for k in K_range:
    km = KMeans(n_clusters=k, init='k-means++', n_init=20, random_state=42)
    labels = km.fit_predict(X_sc)
    results['k'].append(k)
    results['inertia'].append(km.inertia_)
    results['silhouette'].append(silhouette_score(X_sc, labels))
    results['ch'].append(calinski_harabasz_score(X_sc, labels))
    results['db'].append(davies_bouldin_score(X_sc, labels))

res_df = pd.DataFrame(results).set_index('k')

fig, axes = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle('Determinación del Número Óptimo de Clusters — Cuatro Criterios de Validación',
             fontweight='bold', fontsize=13)

configs = [
    ('inertia',    'Inercia (método del codo)',     '#1565C0', 'Minimizar inflexión'),
    ('silhouette', 'Silhouette Score',               '#2E7D32', 'Maximizar'),
    ('ch',         'Calinski-Harabász Index',        '#6A1B9A', 'Maximizar'),
    ('db',         'Davies-Bouldin Index',           '#C62828', 'Minimizar'),
]

for ax, (col, titulo, color, nota) in zip(axes.flat, configs):
    ax.plot(res_df.index, res_df[col], 'o-', color=color, linewidth=2, markersize=7)
    opt = res_df[col].idxmax() if col in ('silhouette', 'ch') else res_df[col].idxmin()
    ax.axvline(opt, color='red', linestyle='--', alpha=0.7, label=f'k={opt}')
    ax.set_xlabel('Número de clusters k')
    ax.set_title(f'{titulo}  ({nota})', fontweight='bold', fontsize=10)
    ax.legend(fontsize=9)
    ax.set_xticks(list(K_range))

plt.tight_layout()
plt.savefig('fig5_optimal_k.png', dpi=150, bbox_inches='tight')
plt.show()
print('\nResumen de métricas por k:')
display(res_df.round(4))
```

### 6.4 DBSCAN — Detección de Outliers Extremos

**DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) identifica puntos de baja densidad (ruido/outliers) sin forzarlos a pertenecer a ningún cluster (Ester et al., 1996).  
El parámetro **ε** se calibra con la curva de k-distancias (k = min_samples) buscando el codo.

```python
# ── Calibración de ε con k-distancias ─────────────────────────────────────────
MIN_SAMPLES = 3
nbrs = NearestNeighbors(n_neighbors=MIN_SAMPLES).fit(X_sc)
distances, _ = nbrs.kneighbors(X_sc)
k_dist = np.sort(distances[:, -1])[::-1]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(k_dist, color='#1565C0', linewidth=2)
ax.set_xlabel('Puntos ordenados por k-distancia')
ax.set_ylabel(f'{MIN_SAMPLES}-NN distancia (ε candidato)')
ax.set_title(f'Curva k-distancias para calibrar ε de DBSCAN (k={MIN_SAMPLES})',
             fontweight='bold')
plt.tight_layout()
plt.savefig('fig8_kdist_dbscan.png', dpi=150, bbox_inches='tight')
plt.show()

# ── Aplicar DBSCAN ────────────────────────────────────────────────────────────
EPS = float(np.percentile(k_dist, 25))  # ajustar visualmente si el codo difiere
db = DBSCAN(eps=EPS, min_samples=MIN_SAMPLES)
df_clu['cluster_db'] = db.fit_predict(X_sc)

n_outliers = (df_clu['cluster_db'] == -1).sum()
n_clusters_db = len(set(df_clu['cluster_db'])) - (1 if -1 in df_clu['cluster_db'].values else 0)
print(f'ε usado: {EPS:.3f}  |  min_samples: {MIN_SAMPLES}')
print(f'Clusters encontrados: {n_clusters_db}')
print(f'Outliers extremos (ruido):  {n_outliers}')

if n_outliers > 0:
    print('\nCooperativas identificadas como outliers extremos:')
    out_coops = df_clu[df_clu['cluster_db'] == -1][['nombre'] + FEATURES_CLU]
    display(out_coops.set_index('nombre').round(2))
```

## 7. Puntuación CAMEL Compuesta

Se construye un **score CAMEL ponderado** siguiendo la metodología de la FDIC/NCUA adaptada para el SFPS (SEPS, 2023).  
Cada indicador se convierte a escala 1–5 según umbrales regulatorios:

| Score | Interpretación |
|---|---|
| **1.0–1.5** | EXCELENTE — solidez alta |
| **1.5–2.5** | BUENO — solidez moderada |
| **2.5–3.5** | ADECUADO — requiere monitoreo |
| **3.5–4.5** | MARGINAL — riesgo elevado |
| **4.5–5.0** | CRÍTICO — riesgo de insolvencia |

Ponderaciones: **C=25%, A=25%, M=20%, E=15%, L=15%** (NCUA, 2023).

```python
def score_breakpoints(valor, breakpoints, scores):
    """Asigna score según puntos de corte; scores tiene len(breakpoints)+1 elementos."""
    if pd.isna(valor): return 3.0
    for bp, sc in zip(breakpoints, scores[:-1]):
        if valor >= bp: return sc
    return scores[-1]

def score_camel(row):
    # C — Capital
    c = score_breakpoints(row.get('C_solvencia', np.nan),
                          [15, 12,  9,  6], [1, 2, 3, 4, 5])

    # A — Asset Quality (media de morosidad y cobertura)
    v_mor = row.get('A_morosidad', np.nan)
    if   pd.isna(v_mor):  a_mor = 3
    elif v_mor < 2:        a_mor = 1
    elif v_mor < 4:        a_mor = 2
    elif v_mor < 6:        a_mor = 3
    elif v_mor < 10:       a_mor = 4
    else:                  a_mor = 5

    v_cob = row.get('A_cobertura', np.nan)
    if   pd.isna(v_cob):  a_cob = 3
    elif v_cob > 200:      a_cob = 1
    elif v_cob > 150:      a_cob = 2
    elif v_cob > 100:      a_cob = 3
    elif v_cob > 75:       a_cob = 4
    else:                  a_cob = 5

    a = (a_mor + a_cob) / 2

    # M — Management
    v_ef = row.get('M_efic_operativa', np.nan)
    if   pd.isna(v_ef):   m = 3
    elif v_ef < 3:         m = 1
    elif v_ef < 4:         m = 2
    elif v_ef < 5.5:       m = 3
    elif v_ef < 7:         m = 4
    else:                  m = 5

    # E — Earnings
    v_roa = row.get('E_roa', np.nan)
    if   pd.isna(v_roa):  e = 3
    elif v_roa > 1.5:      e = 1
    elif v_roa > 0.75:     e = 2
    elif v_roa > 0:        e = 3
    elif v_roa > -0.5:     e = 4
    else:                  e = 5

    # L — Liquidity
    v_liq = row.get('L_liq_ampliada', np.nan)
    if   pd.isna(v_liq):  l = 3
    elif v_liq > 25:       l = 1
    elif v_liq > 20:       l = 2
    elif v_liq > 14:       l = 3
    elif v_liq > 10:       l = 4
    else:                  l = 5

    camel = 0.25*c + 0.25*a + 0.20*m + 0.15*e + 0.15*l
    return pd.Series({'sc_C': c, 'sc_A': a, 'sc_M': m, 'sc_E': e, 'sc_L': l, 'camel': camel})

scores = df.apply(score_camel, axis=1)
df = pd.concat([df, scores], axis=1)

def rating_camel(v):
    if v <= 1.5: return 'EXCELENTE'
    if v <= 2.5: return 'BUENO'
    if v <= 3.5: return 'ADECUADO'
    if v <= 4.5: return 'MARGINAL'
    return 'CRÍTICO'

df['rating_camel'] = df['camel'].apply(rating_camel)

COLORES_RATING = {
    'EXCELENTE': '#1B5E20', 'BUENO': '#4CAF50',
    'ADECUADO': '#FF9800',  'MARGINAL': '#E65100', 'CRÍTICO': '#B71C1C'
}

print('\nDistribución de ratings CAMEL:')
print(df['rating_camel'].value_counts().to_string())
```

```python
cols_ranking = ['nombre','activo_total',
                'C_solvencia','A_morosidad','A_cobertura',
                'M_efic_operativa','E_roa','L_liq_ampliada',
                'sc_C','sc_A','sc_M','sc_E','sc_L','camel','rating_camel','cluster_km']

ranking = (df[cols_ranking]
           .sort_values('camel')
           .assign(activo_m=lambda d: (d.activo_total/1e6).round(1))
           .drop(columns='activo_total'))

def color_rating(val):
    colores_bg = {
        'EXCELENTE':'#C8E6C9','BUENO':'#DCEDC8',
        'ADECUADO':'#FFE0B2','MARGINAL':'#FFCCBC','CRÍTICO':'#FFCDD2'
    }
    return f'background-color: {colores_bg.get(val,"")}'

def color_camel_num(val):
    if val <= 1.5: bg = '#C8E6C9'
    elif val <= 2.5: bg = '#DCEDC8'
    elif val <= 3.5: bg = '#FFE0B2'
    elif val <= 4.5: bg = '#FFCCBC'
    else: bg = '#FFCDD2'
    return f'background-color: {bg}'

display(
    ranking.style
    .map(color_rating, subset=['rating_camel'])
    .map(color_camel_num, subset=['camel'])
    .format({'C_solvencia':'{:.1f}%','A_morosidad':'{:.2f}%',
             'A_cobertura':'{:.1f}%','M_efic_operativa':'{:.2f}%',
             'E_roa':'{:.3f}%','L_liq_ampliada':'{:.1f}%',
             'camel':'{:.3f}','activo_m':'${:.0f}M'},
            na_rep='N/D')
)
```

```python
# Extraer serie histórica anual (diciembre de cada año) para todo Segmento 1
C_NODEV_SQL = "'1425','1426','1427','1428','1432','1433','1434','1435','1436','1440','1441','1442','1443','1444','1448'"
C_VENC_SQL  = "'1449','1450','1451','1452','1456','1457','1458','1459'"

hist = con.execute(f"""
    SELECT
        MONTH(fecha_corte)  AS mes,
        fecha_corte,
        COUNT(DISTINCT ruc) AS n_coops,
        SUM(CASE WHEN cuenta='1'  THEN saldo_usd END) / 1e6  AS activo_total_m,
        SUM(CASE WHEN cuenta='2'  THEN saldo_usd END) / 1e6  AS pasivo_m,
        SUM(CASE WHEN cuenta='3'  THEN saldo_usd END) / 1e6  AS patrimonio_m,
        SUM(CASE WHEN cuenta='14' THEN saldo_usd END) / 1e6  AS cartera_m,
        SUM(CASE WHEN cuenta='21' THEN saldo_usd END) / 1e6  AS depositos_m,
        SUM(CASE WHEN cuenta='1499' THEN saldo_usd END) / 1e6 AS provisiones_m,
        SUM(CASE WHEN cuenta='11' THEN saldo_usd END) / 1e6  AS fondos_m,
        SUM(CASE WHEN cuenta='13' THEN saldo_usd END) / 1e6  AS inver_m,
        SUM(CASE WHEN cuenta='4'  THEN saldo_usd END) / 1e6  AS gastos_m,
        SUM(CASE WHEN cuenta='5'  THEN saldo_usd END) / 1e6  AS ingresos_m,
        SUM(CASE WHEN cuenta='41' THEN saldo_usd END) / 1e6  AS int_causados_m,
        SUM(CASE WHEN cuenta='45' THEN saldo_usd END) / 1e6  AS gastos_op_m,
        SUM(CASE WHEN cuenta='51' THEN saldo_usd END) / 1e6  AS int_ganados_m,
        SUM(CASE WHEN cuenta IN ({C_NODEV_SQL}) THEN saldo_usd ELSE 0 END)/1e6 AS nodev_m,
        SUM(CASE WHEN cuenta IN ({C_VENC_SQL})  THEN saldo_usd ELSE 0 END)/1e6 AS vencida_m
    FROM eeff_mensuales
    WHERE segmento = 'SEGMENTO 1'
      AND YEAR(fecha_corte) >= 2018
      AND razon_social NOT LIKE '%LIQUIDACI%'
    GROUP BY fecha_corte
    ORDER BY fecha_corte
""").df()

# Calcular indicadores del sistema (datos agregados)
hist['resultado_m']     = hist['ingresos_m'] - hist['gastos_m']
hist['cart_improd_m']   = hist['nodev_m'] + hist['vencida_m']
hist['sist_solvencia']  = hist['patrimonio_m'] / hist['activo_total_m'] * 100
hist['sist_morosidad']  = hist['cart_improd_m'] / hist['cartera_m'] * 100
hist['sist_cobertura']  = hist['provisiones_m'].abs() / hist['cart_improd_m'] * 100
hist['sist_roa']        = hist['resultado_m'] / hist['activo_total_m'] * 100
hist['sist_liq_amp']    = (hist['fondos_m'] + hist['inver_m']) / hist['depositos_m'] * 100
hist['sist_nim']        = (hist['int_ganados_m'] - hist['int_causados_m']) / hist['activo_total_m'] * 100
hist['sist_efic_op']    = hist['gastos_op_m'] / hist['activo_total_m'] * 100

print(f'Períodos históricos extraídos: {len(hist)}')
print(f'Desde: {hist.fecha_corte.min().date()}  →  Hasta: {hist.fecha_corte.max().date()}')
```

## 10. Análisis Estadístico y Econométrico

### 10.1 Test de Kruskal-Wallis entre Clusters

Se aplica el test no paramétrico de **Kruskal-Wallis** para verificar si los clusters difieren significativamente en cada indicador (Kruskal & Wallis, 1952).  
$H_0$: las medianas son iguales en todos los clusters. Se rechaza si $p < 0.05$.

```python
from scipy.stats import kruskal, mannwhitneyu
from itertools import combinations

print('=' * 60)
print('TEST DE KRUSKAL-WALLIS POR INDICADOR (H0: medianas iguales)')
print('=' * 60)

kw_results = []
for r in RATIOS:
    grupos = [df[df['cluster_km']==cl][r].dropna().values
              for cl in sorted(df['cluster_km'].dropna().unique())]
    grupos = [g for g in grupos if len(g) >= 2]
    if len(grupos) < 2: continue
    stat, pval = kruskal(*grupos)
    kw_results.append({
        'Indicador': r, 'H-stat': round(stat,3),
        'p-value': round(pval, 4),
        'Significativo': '✓' if pval < 0.05 else ''
    })

kw_df = pd.DataFrame(kw_results).set_index('Indicador').sort_values('p-value')
display(kw_df)
print(f"\nIndicadores con diferencias significativas entre clusters (p<0.05): "
      f"{(kw_df['p-value'] < 0.05).sum()} de {len(kw_df)}")
```

### 10.2 Regresión OLS — Determinantes de la Capitalización

Se estima un modelo de regresión OLS para identificar los factores asociados con la solvencia técnica (capitalización). Se incluye el logaritmo del activo como variable de control de tamaño.

```python
df_ols = df[['C_solvencia','A_morosidad','M_efic_operativa',
             'E_roa','L_liq_ampliada','S_aprovision',
             'log_activo']].dropna().copy()

formula = ('C_solvencia ~ A_morosidad + M_efic_operativa + '
           'E_roa + L_liq_ampliada + S_aprovision + log_activo')

modelo = smf.ols(formula, data=df_ols).fit(cov_type='HC3')  # errores robustos
print(modelo.summary())

# Test de normalidad de residuos (Shapiro-Wilk)
stat_sw, p_sw = shapiro(modelo.resid)
print(f'\nShapiro-Wilk residuos: stat={stat_sw:.4f}, p={p_sw:.4f}')
print('Residuos normales' if p_sw > 0.05 else 'Residuos NO normales — interpretar con cautela')

# Visualización de residuos
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(modelo.fittedvalues, modelo.resid, alpha=0.7, color='#1565C0')
axes[0].axhline(0, color='red', linestyle='--')
axes[0].set_xlabel('Valores ajustados'); axes[0].set_ylabel('Residuos')
axes[0].set_title('Residuos vs Valores Ajustados', fontweight='bold')

from scipy.stats import probplot
probplot(modelo.resid, plot=axes[1])
axes[1].set_title('Q-Q Plot de Residuos', fontweight='bold')

plt.tight_layout()
plt.savefig('fig13_ols_residuos.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Correlaciones de Spearman con la capitalización (ranking por |rho|) ────────
print('Correlaciones de Spearman con C_solvencia:')
rho_rows = []
for r in [x for x in RATIOS if x != 'C_solvencia']:
    s1 = df[['C_solvencia', r]].dropna()
    if len(s1) < 5: continue
    rho, p = spearmanr(s1['C_solvencia'], s1[r])
    rho_rows.append({'Indicador': r, 'ρ Spearman': round(rho,4),
                     'p-valor': round(p,4),
                     'Sig.': '***' if p<0.001 else '**' if p<0.01 else '*' if p<0.05 else ''})

rho_df = pd.DataFrame(rho_rows).set_index('Indicador').sort_values('ρ Spearman')
display(rho_df)
```
